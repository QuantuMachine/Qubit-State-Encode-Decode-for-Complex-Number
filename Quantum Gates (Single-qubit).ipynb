# **Qiskit call (or import).**



````python
'''
   Import Quantum register, circuit etc//
'''
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, Aer, assemble
from qiskit.visualization import plot_histogram
sim = Aer.get_backend('aer_simulator')

'''
   To import the general libraries//
'''
import numpy as np

'''
    Define registers & circuit//
'''
q = QuantumRegister(1);
c = ClassicalRegister(1);
````

```
 ############
```
# **Ex_1.1:- Qubit built-in initial state.**

````python
'''Part. 1: - Circuit'''
circuit_initial = QuantumCircuit(q, c);
circuit_initial.measure(q, c);
print(circuit_initial); 

'''Part. 2: - Histogram//'''
qobj = assemble(circuit_initial)  # To assemble the circuit to run it//
counts = sim.run(qobj).result().get_counts()  # For simulation and to return the state vector//
plot_histogram(counts)  # To display result//
````

````python
circuit_initial.draw(output='mpl', style={'backgroundcolor': '#EEEEEE'})
````
